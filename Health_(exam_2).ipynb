# **7. Basic Line/Bar Grap**

```python
# Line Graph the homicides in brasil 2017-2021
fig = px.line(brasil, x = 'Year_str', y='Hom')


# Change title
fig.update_layout(title = 'Homicides in Brasil 2017-2021',
                  title_font_size = 30,
                  title_x = 0.5)

# Change Axis
fig.update_xaxes(title = 'Years', title_font_size = 20)
fig.update_yaxes(title = 'Homicides', title_font_size = 20)

fig.show()
```

```python
# Bar Graph the homicides in brasil 2017-2021

fig = px.bar(brasil, x = "Year_str", y='Hom')

# Change title
fig.update_layout(title = 'Homicides in Brasil 2017-2021',
                  title_font_size = 30,
                  title_x = 0.5,
                  yaxis_range = [1500,2500])

# Change Axis
fig.update_xaxes(title = 'Homicides', title_font_size = 20)
fig.update_yaxes(title = 'Years', title_font_size = 20)

fig.show()
```

## Two y-axis
