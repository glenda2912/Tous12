# **3.Cleaning**

## Column Names 

```python
# Change Time for Year 
d.rename(columns = {"Años": "Year"}, inplace = True)

# Change County Name
d.rename(columns = {"País": "Country"}, inplace = True)

#Change Value
d.rename(columns = {"Valor": "Hom"}, inplace = True)

#Check
d.head()
```

## A few more things

```python
# (Optional) Duplicate the dataframe

pp = d
```

```python
# Duplicate year variable 
pp["Year_str"] = pp["Year"].astype(str)

#Check
pp.head()
```

# **4.Indexing**

```python
# Create an index
idx = pd.MultiIndex.from_product([pp['Country'].unique(),
                                  range(pp.Year.min(),pp.Year.max()+1)],
                                  names= ['Country', 'Year'])

#Reindex the dataframe 
pp = pp.set_index(['Year','Country'])

#Check 
pp.head()                
```

# **5.Basic Statistics**

```python
# By variable 
pp[['Hom']].describe()
```

```python
# By variable and by year 
pp[['Hom']].groupby('Year').describe()
```

```python
# By variable and by country
pp[['Hom']].groupby('Country').describe()
```

# **6. Filters and pivots**


## Simple Filters 

```python
# Filter by country name and save it 
brasil = pp.filter(like= 'Brasil', axis=0)
print(brasil)
```

```python
# Filter by year name and save it 
y20 = pp.filter(like= '2019', axis=0)
print(y20)
```

## Simple Pivots

```python
# Creating is dataframe with the tie-serie of hom 
ts_Hom = d.pivot(index   = 'Year',
                 columns = 'Country',
                 values  = 'Hom')
print(ts_Hom)
```
